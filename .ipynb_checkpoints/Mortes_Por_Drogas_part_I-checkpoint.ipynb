# Ednei Cunha Vicente

 O conjunto de dados de mortes devido a drogas em Connecticut contém informações sobre 5106 pessoas que morreram devido a overdose de drogas entre 2012 e 2018 em Connecticut, EUA.
 

O conjunto de dados inclui dados relacionados à idade, raça, sexo, local de residência das vítimas, bem como as drogas que tomaram por overdose. Essas informações podem ser usadas para entender se o uso de drogas é prevalente em uma área ou cidade específica, o uso de drogas por indivíduos de diferentes faixas etárias e raças, bem como a popularidade de diferentes tipos de drogas.

Este conjunto de dados foi obtido do [governo dos EUA
Open Data Initiative] (https://data.gov) [Conjunto de dados de mortes relacionadas a drogas acidentais] (https://catalog.data.gov/dataset/accidental-drug-related-deaths-january-2012-sept-2015).
A página Open Data Initiative menciona o seguinte como a fonte original do
conjunto de dados:
* Governo local, Connecticut 

 O objetivo desta analise sera responder perguntas sobre o impacto que as droga tiveram em tais óbitos.  

Criar um modelo de cluster usando o algoritmo k_means,  PCA para redução de dimensionalidade e a métrica silhuette_score.

Algumas informacoes sobre o estado de Connecticut:
    Connecticut (ou Coneticute[2] ou Conecticute) é um dos 50 estados dos Estados Unidos, localizado na região da 
    Nova Inglaterra. Connecticut é o terceiro menor estado norte-americano em extensão territorial; somente Delaware e 
    Rhode Island são menores.
    Connecticut é o quarto estado norte-americano mais densamente povoado do país.
    A principal fonte de renda de Connecticut é a prestação de serviços financeiros e imobiliários.

```python
#Dando nome as colunas para facilitar a analise
cols = ['Data','Idade','Sexo','Etnia','Cidade','Condado','Morte/Cidade','Morte/Condado','Localizacao','Decricao lesao',
       'Local lesao','Cidade/lesao','Condado/Lesao','Heroina','Cocaina','Fantanil','FantanilAnalogue','Oxicodona','Oximorfona',
       'Etanol','Hydrocodone','Benzodiazepina','Metadona','Anfeto','Tramad','Morfine_NtoHeroine','Hidromorfina','Outros',
       'OpiaceosNOS','AnyOpioid','Maneira/Morte','Cidade/Morte','Residencia/Morte','InjuryCityGeo']
```

```python
#carregando os dados
df = pd.read_csv('MortesPorDrogasConnecticut-2012-2018.csv')
```

```python
#Excluindo registro que nao serao utilizados
to_drop = ['ID','DateType','LocationifOther','OtherSignifican','InjuryState','ResidenceState','COD']
df.drop(to_drop,inplace= True , axis =1)
```

```python
#checando o percentual de valores missing
missing = df.isnull().sum().sort_values(ascending=False)
missing = missing[missing > 0]/df.shape[0]
print(f'{missing * 100}%')
```

```python
#funcao para preencher o valor NAN da coluna idade
def imput_media(df):
    return df.fillna(df.mean())
df['Idade'] = df['Idade'].transform(imput_media)
print(df['Idade'].isnull().sum())
```

```python
#alterando o formato da coluna data
df.Data = pd.to_datetime(df['Data'])
df["AnoMes"] = df["Data"].map(lambda date : 100 * date.year + date.month)

#Alterando o label das colunas "drogas" com '0' 'y'sim, '1' NaN 'nao'
df[['Heroina','Cocaina','Fantanil','FantanilAnalogue','Oxicodona','Oximorfona','Etanol','Hydrocodone',
                 'Benzodiazepina','Metadona','Anfeto','Tramad','Morfine_NtoHeroine','Hidromorfina','Outros','OpiaceosNOS',
                'AnyOpioid']] = df[['Heroina','Cocaina','Fantanil','FantanilAnalogue','Oxicodona','Oximorfona','Etanol','Hydrocodone',
                 'Benzodiazepina','Metadona','Anfeto','Tramad','Morfine_NtoHeroine','Hidromorfina','Outros','OpiaceosNOS',
                'AnyOpioid']].apply(pd.Series.replace, to_replace = 'Y', value = 0)
df[['Heroina','Cocaina','Fantanil','FantanilAnalogue','Oxicodona','Oximorfona','Etanol','Hydrocodone',
                 'Benzodiazepina','Metadona','Anfeto','Tramad','Morfine_NtoHeroine','Hidromorfina','Outros','OpiaceosNOS',
                'AnyOpioid']] = df[['Heroina','Cocaina','Fantanil','FantanilAnalogue','Oxicodona','Oximorfona','Etanol','Hydrocodone',
                 'Benzodiazepina','Metadona','Anfeto','Tramad','Morfine_NtoHeroine','Hidromorfina','Outros','OpiaceosNOS',
                'AnyOpioid']].fillna(1)

#Alterando algumas colunas que restaram
df['Fantanil'].replace(to_replace=['Y-A','Y (PTCH)','Y POPS'],value=1 , inplace=True)
df['Morfine_NtoHeroine'].replace(to_replace=['PCP NEG','NO RX BUT STRAWS','YES','STOLE MEDS'],value=1 , inplace=True)
df['AnyOpioid'].replace(to_replace=['N'],value=1 , inplace=True)

#deletando a coluna "outros"
del df["Outros"]

#Alterando a coluna Etnia
df['Etnia'] = df['Etnia'].str.replace('Hispanic, White','Hispanic')
df['Etnia'] = df['Etnia'].str.replace('Hispanic, Black', 'Hispanic')
df['Etnia'].replace(to_replace=['Unknown','Asian, Other','Asian Indian','Other','Chinese','Native American, Other','Hawaiian'],value='Other',inplace=True)
```

```python
#Alterando valores nulos da coluna "Etnia"
nov_etnia = np.where(df["Etnia"].isnull(),'Other',df["Etnia"])
df["Etnia"] = nov_etnia
df["Etnia"].value_counts().plot(kind='bar')
```

```python
df["Etnia"].isnull().sum()
```

```python
#Alterando valores nulos da coluna "Sexo"
nov_sexo = np.where(df["Sexo"].isnull(),'Female',df["Sexo"])
df["Sexo"] = nov_sexo 
df = df[df['Sexo']!= 'Unknown']
df["Sexo"].value_counts().plot(kind='bar')
```

```python
df['Sexo'].isnull().sum()
```

# Criando graficos de variáveis únicas 

```python
def cria_histograma(df, features, rows, cols):
    fig = plt.figure(figsize = (20,20))
    
    for i, feature in enumerate(features):
        ax = fig.add_subplot(rows, cols, i+1)
        df[feature].hist(bins = 20, ax = ax, facecolor = 'red')
        ax.set_title(feature , color = 'Red')
        
    fig.tight_layout()  
    plt.show()
```

```python
# Executa a função
cria_histograma(novo_df, novo_df.columns, 7, 3) #"0" representa sim , "1"representao nao
```

Podemos ver que vários tipos de drogas existem nesse conjunto de dados, vamos criar uma variável que contenha a quantidade drogas usadas por cada pessoa.

```python
novo_df.columns
```

```python
drogas = [ 'Heroina',
       'Cocaina', 'Fantanil', 'FantanilAnalogue', 'Oxicodona', 'Oximorfona',
       'Etanol', 'Hydrocodone', 'Benzodiazepina', 'Metadona', 'Anfeto',
       'Tramad', 'Morfine_NtoHeroine', 'Hidromorfina', 'OpiaceosNOS',
       'AnyOpioid']
```

```python
for drog in drogas:
    if drog in novo_df.columns:
        colname = str(drog) + 'temp'
        novo_df[colname] = novo_df[drog].apply(lambda x: 0 if x=='Nao' else 1)
```

```python
novo_df.head()
```

```python
novo_df['ContagemDrogas'] = 0
```

```python
for drog in drogas:
    if drog in novo_df.columns:
        colname = str(drog) + 'temp'
        novo_df['ContagemDrogas'] = novo_df['ContagemDrogas'] + novo_df[colname]
        del novo_df[colname]
```

Com o comando "value_counts" é possivel notar que a maioria de óbitos são com 2 e 3 tipos de drogas.

```python
novo_df['ContagemDrogas'].value_counts()
```
